# Obtener Imagenes desde Amazon

```python
images = []
sufix = "https://m.media-amazon.com"
imgs = soup.find("div",{"class":"s-main-slot s-result-list s-search-results sg-row"}).findAll("img")
#img = soup.find("div",{"class":"a-section aok-relative s-image-square-aspect"}).find("img")
for aux in imgs:
    if (aux["src"].find(sufix)!=-1):
        images.append(aux["src"])
#print(images.length)
len(images)
```

```python
images = list(dict.fromkeys(images))
len(images)
```

# Descargar imagenes

```python
i=0
for url in images:
    imagen = requests.get(url).content
    with open(str(i)+".jpg", 'wb') as handler:
        handler.write(imagen)
    i+=1
```
